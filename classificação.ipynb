## Dados categóricos X situacao (retrabalhado)

Consideração: O número de classes no target é muito alto, o que acaba atrapalhando a interpretação dos dados visualmente. 

Pensando nisso, no desbalanceamento dos dados,na dificuldade dos modelos de L.M. para interpretar esses dados, e o objetivo deste trabalho em entender as causas de desistência nos cursos de BCC e ADS daa UTFPR, será proposto na próxima seção a discretização da situação em apenas 3 classes: "Formado", "Desistente", "Regular" e "Outros".

```python
# Processamento para deixar apenas Formado, Desistente, Regular e o resto vira Outros
condicao = relacao_alunos['situacao'].isin(['Formado', 'Desistente', 'Regular'])

relacao_alunos['situacao'] = relacao_alunos['situacao'].where(condicao, 'Outros')
relacao_alunos['situacao'].unique().tolist()
```

### curso x situação

```python
fig, ax = plt.subplots(1,2, figsize=(32,4))
plt.figure(figsize=(16,4))
sns.countplot(relacao_alunos, x='curso', hue='situacao', ax=ax[0])
sns.countplot(relacao_alunos, x='situacao', hue='curso',  ax=ax[1])
fig.show();
```

### forma de ingresso x situação

```python
fig, ax = plt.subplots(1,2, figsize=(32,4))
plt.figure(figsize=(16,4))
sns.countplot(relacao_alunos, x='forma_de_ingresso', hue='situacao', ax=ax[0])
sns.countplot(relacao_alunos, x='situacao', hue='forma_de_ingresso',  ax=ax[1])
fig.show();
```

```python
# por estes gráficos é possível notar que não há impacto numérico expressivo da forma de ingresso na situação para mantermos o grande numero destas classes
# mudança proposta: SISU, Vestibular, Outros

fig, ax = plt.subplots(1,2, figsize=(64,4))
plt.figure(figsize=(16,4))
sns.countplot(relacao_alunos.query('situacao != "Desistente" and forma_de_ingresso != "SISU - Sistema de Seleção Unificada" and forma_de_ingresso != "Vestibular"'), x='forma_de_ingresso', hue='situacao', ax=ax[0])
sns.countplot(relacao_alunos.query('situacao == "Desistente" and forma_de_ingresso != "SISU - Sistema de Seleção Unificada" and forma_de_ingresso != "Vestibular"'), x='situacao', hue='forma_de_ingresso',  ax=ax[1])
fig.show();
```

```python
# Por estes gráficos, é notável que por mais que existam muitos alunos que entraram pelo SISU, há uma grande representatividade no número de desistentes desta classe em comparação ao vestibular
# Outro ponto relevante foi o período que a UTFPR não realizou vestibular interno, o que é perceptível ao analisarmos a coluna Regular em Vestibular no gráfico da esquerda

fig, ax = plt.subplots(1,2, figsize=(32,4))
plt.figure(figsize=(16,4))
sns.countplot(relacao_alunos.query('situacao != "Desistente" and (forma_de_ingresso == "SISU - Sistema de Seleção Unificada" or forma_de_ingresso == "Vestibular")'), x='forma_de_ingresso', hue='situacao', ax=ax[0])
sns.countplot(relacao_alunos.query('situacao == "Desistente" and (forma_de_ingresso == "SISU - Sistema de Seleção Unificada" or forma_de_ingresso == "Vestibular")'), x='situacao', hue='forma_de_ingresso',  ax=ax[1])
fig.show();
```

```python
condicao = relacao_alunos['forma_de_ingresso'].isin(['SISU - Sistema de Seleção Unificada', 'Vestibular'])

relacao_alunos['forma_de_ingresso'] = relacao_alunos['forma_de_ingresso'].where(condicao, 'Outros')
relacao_alunos['forma_de_ingresso'].value_counts()
```

```python
# De fato a classe "outros" da forma_de_ingresso é pouco expressiva ao analisarmos as disistências
# SISU ser o mais expressivo se dá pela possibilidade de ingresso durante um período ter se dado apenas por esta modalidade

fig, ax = plt.subplots(1,2, figsize=(16,4))
plt.figure(figsize=(16,4))
sns.countplot(relacao_alunos, x='forma_de_ingresso', hue='situacao', ax=ax[0])
sns.countplot(relacao_alunos, x='situacao', hue='forma_de_ingresso',  ax=ax[1])
fig.show();
```

### cota x situacao

```python
plt.figure(figsize=(32,4))
sns.countplot(relacao_alunos, x='sigla_cota', hue='situacao')
```

```python
# Pelos gráficos é possível notar que a descentralização da informação de ser cotista pode impactar na analise quando visualizamos os desistentes
# como os dados predominantes são sem a caracterização da categoria da cota, uma simplificação pode ser proveitosa tanto para análise quanto para os modelos de L.M.
# mudança proposta: cotista, não cotista

plt.figure(figsize=(32,4))
sns.countplot(relacao_alunos, x='situacao', hue='sigla_cota')
```

```python
condicao = relacao_alunos['sigla_cota'].isin(['Cotista', 'Não cotista', 'Sem cotas'])
relacao_alunos['sigla_cota'] = relacao_alunos['sigla_cota'].where(condicao, 'Cotista')

condicao = relacao_alunos['sigla_cota'].isin(['Cotista', 'Não cotista'])
relacao_alunos['sigla_cota'] = relacao_alunos['sigla_cota'].where(condicao, 'Não cotista')

relacao_alunos['sigla_cota'].value_counts()
```

```python
# Por estes gráficos é possível notarmos que alunos não cotistas aparentam ter maior chance de se tornarem desistentes

fig, ax = plt.subplots(1,2, figsize=(16,4))
plt.figure(figsize=(16,4))
sns.countplot(relacao_alunos, x='sigla_cota', hue='situacao', ax=ax[0])
sns.countplot(relacao_alunos, x='situacao', hue='sigla_cota',  ax=ax[1])
fig.show();
```
